## Clean Data
This dataset relies on nursing homes to self-report their availability on a weekly basis. To double-check there's not any null or N/A values in important columns, I used DropNa.

```python
clean_beds = beds.dropna(subset=["Facility Name", "Certification Number", "County", "Bed Type", "Week Difference Category", "Week Difference", "Total Capacity", "Available Capacity"])
```

## Pare Down Data: Dutchess County
For my story, I'm trying to add up the availability specifically in Dutchess County. 

```python
county_beds = beds[["County", "Facility Name", "Bed Type", "Week Difference", "Available Capacity", "Total Capacity"]]
county_beds
```

```python
dutchess_beds = county_beds.loc[county_beds["County"] == "Dutchess"]
dutchess_beds
```

Ugh...so even though row 52 isn't technically empty, it's value is only a "." This will be counted as null if I use sum(), but we need to change the dtype of "Available Capacity" first, and this is getting in the way. I can't figure out how to do it to the larger data set as a whole, so for now we're going to take the long way around and address any "." as we go.

```python
dutchess_beds.loc[52, "Available Capacity"] = 0
dutchess_beds
```

```python
dutchess_beds["Available Capacity"] = dutchess_beds["Available Capacity"].astype(int)
```

```python
dutchess_beds["Available Capacity"].sum()
```

```python
dutchess_beds["Total Capacity"].sum()
```

```python
dutchess_vacancy = (328/2072)*100 #Percentage of beds vacant in Dutchess County
```

Let's put the vacancies in their own dataframe, so I can export it and make something in Datawrapper. 

```python
data = {"Location": ["Statewide", "Columbia County", "Dutchess County", "Greene County", "Orange County", "Putnam County", "Rockland County", "Sullivan County", "Ulster County", "Westchester County"],
       "Occupied Beds": [(100-total_vacancy), (100-columbia_vacancy), (100-dutchess_vacancy), (100-greene_vacancy), (100-orange_vacancy), (100-putnam_vacancy), (100-rockland_vacancy), (100-sullivan_vacancy), (100-ulster_vacancy), (100-westchester_vacancy)],
       "Vacancy Rate": [total_vacancy, columbia_vacancy, dutchess_vacancy, greene_vacancy, orange_vacancy, putnam_vacancy, rockland_vacancy, sullivan_vacancy, ulster_vacancy, westchester_vacancy]
       }
df = pd.DataFrame(data)
```

```python
df
```

```python
df["Vacancy Rate"] = df["Vacancy Rate"].round(2) 
df["Occupied Beds"] = df["Occupied Beds"].round(2) #Those percentages are too long, let's round them. 
```
